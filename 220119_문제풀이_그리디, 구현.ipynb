### 거스름돈
https://www.acmicpc.net/problem/14916

```python
n = int(input())

x = n//5
target = x*5

if (n - target) % 2 == 1:
    c5 = x-1
    c2 = (n - 5*(x-1)) // 2
else:
    c5 = x
    c2 = (n - 5*x) // 2

if c5 < 0 or c2 < 0:
    print(-1)
else:
    print(c5+c2)
```

### ATM
https://www.acmicpc.net/problem/11399

```python
n = int(input())
p = list(map(int, input().split()))

p.sort()
# print(p)

sum = 0
for i in range(len(p)):
    sum += (5-i)*p[i]

print(sum)
```
